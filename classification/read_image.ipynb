```python
# 帮我把文件夹下的with_shortcut.png和without_shortcut.png两个图片读取并且并列显示，然后下标分别是with_shortcut和without_shortcut
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# 读取图片
img_with_shortcut = mpimg.imread('with_shortcut.png')
img_without_shortcut = mpimg.imread('without_shortcut.png')

# 创建一个并排的图形
fig, ax = plt.subplots(1, 2, figsize=(16, 8))

# 显示第一个图片
ax[0].imshow(img_with_shortcut)
ax[0].set_title('with_shortcut')
ax[0].axis('off')  # 隐藏坐标轴

# 显示第二个图片
ax[1].imshow(img_without_shortcut)
ax[1].set_title('without_shortcut')
ax[1].axis('off')  # 隐藏坐标轴

# 调整布局以防止标题被裁剪
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

# 获取 misclassified 文件夹中的前9个图片文件名
image_folder = 'misclassified'
image_files = sorted(os.listdir(image_folder))[:6]  # 获取前9个图片文件名

# 创建一个3x3的图形
fig, axes = plt.subplots(2, 3, figsize=(9, 6))

# 遍历9张图片并显示
for i, ax in enumerate(axes.flat):
    img_path = os.path.join(image_folder, image_files[i])
    img = mpimg.imread(img_path)
    ax.imshow(img)
    # ax.set_title(f'Image {i+1}')  # 可根据实际需要修改标题
    ax.axis('off')  # 隐藏坐标轴

# 调整布局
plt.tight_layout()
plt.show()
```
